#### (2) 상장법인목록 - crawling from the krx site 
http://kind.krx.co.kr/corpgeneral/corpList.do?method=loadInitPage

```python
import pandas as pd
import numpy as np
import requests
from io import BytesIO

url = 'http://kind.krx.co.kr/corpgeneral/corpList.do'
data = {
    'method' : 'download' , # corpList 를 download 로 바꾸어줘야 됨 
    'orderMode' : '3' , # 정렬컬럼
    'orderStat' : 'D' , # Descending order 
    'searchType' : '13' , # 검색 유형 : 상장법인 
    'fiscalYearEnd' : 'all' ,  
    'location': 'all'
}

r = requests.post(url, data = data) 
f = BytesIO(r.content) 
dfs = pd.read_html(f, header=0)

dfs = dfs[0][["회사명" , "종목코드" , "업종" , "주요제품" , "상장일"]]
dfs["종목코드"] =  dfs["종목코드"].astype(np.str)
```

```python
# dfs data type 바뀐여부 체크하기 
dfs.dtypes
```

```python
# dfs 보기
dfs.tail( n = 6)
```

```python
%matplotlib inline

dfp = dfs[['회사명', '상장일']]
dfp.columns = ['Corps', 'Date'] 
dfp.groupby([dfp['Date'],pd.TimeGrouper(freq='A')])
```

```python
dfp = dfp.set_index('Date')
dfp = dfp.resample('M').count()
dfp.plot(kind = "hist")
```
